```python
import pandas as pd
```

```python
df=pd.read_csv("D:\Data science\machine learning\Feature Scalling\Label Encoding\shop data.csv")
df
```

```python
df['income'].unique()
```

```python
#it assugn value orderly, like low=0,medium=1,high=2
income_order=[ 'low','medium','high',]
```

```python
from sklearn.preprocessing import OrdinalEncoder
enco=OrdinalEncoder(categories=[income_order])
enco.fit(df[['income']])
enco_data=enco.transform(df[['income']])
final=pd.DataFrame(data=enco_data,columns=['income'])
final
```
